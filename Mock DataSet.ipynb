```python
from path import Path
import pandas as pd
```

```python
data = Path('jantojun2020.csv')
df = pd.read_csv(data)
df.head(25)
```

```python
clean_df=df.fillna(0)
```

```python
clean_df.dtypes
```

```python
clean_df.head(25)
```

```python
filtered_df= clean_df.drop(['YEAR', 'QUARTER', 'TAIL_NUM','ORIGIN','ORIGIN_CITY_NAME','ORIGIN_STATE_ABR', 'DEST', 'DEST_CITY_NAME','DEST_STATE_ABR', 'CRS_DEP_TIME', 'DEP_TIME','DEP_DELAY_NEW','DEP_DEL15','DEP_DELAY_GROUP','DEP_TIME_BLK','TAXI_OUT','WHEELS_OFF','WHEELS_ON','TAXI_IN','CRS_ARR_TIME','ARR_TIME', 'FL_DATE','CRS_ELAPSED_TIME','ACTUAL_ELAPSED_TIME','AIR_TIME', 'DISTANCE_GROUP', 'ARR_TIME_BLK'],axis=1)
```

```python
filtered_df
```

```python
filtered_df.dtypes
```

```python
Origin_State = pd.unique(filtered_df['ORIGIN_STATE_NM'])
```

```python
Origin_State
```

```python
state_counts = filtered_df.ORIGIN_STATE_NM.value_counts()
state_counts
```

```python
state_counts.plot.density()
```

```python
# Determine which values to replace
replace_states = list(state_counts[state_counts < 100000].index)

# Replace in DataFrame
for state in replace_states:
    filtered_df.ORIGIN_STATE_NM = filtered_df.ORIGIN_STATE_NM.replace(state,"Other")


# Check to make sure binning was successful
filtered_df.ORIGIN_STATE_NM.value_counts()
```

```python
# Create the OneHotEncoder instance
from sklearn.preprocessing import OneHotEncoder
enc= OneHotEncoder(sparse=False)

#fit the encoder and the produced encoded DataFrame
encoded_plane_df=pd.DataFrame(enc.fit_transform(filtered_df.ORIGIN_STATE_NM.values.reshape(-1,1)))

#Rename encoded Columns
encoded_plane_df.columns=enc.get_feature_names(['ORIGIN_STATE_NM'])
encoded_plane_df.head()
```

```python
dest_state_counts = filtered_df.DEST_STATE_NM.value_counts()
dest_state_counts
```

```python
# Determine which values to replace
replace_dest_states = list(dest_state_counts[dest_state_counts < 100000].index)

# Replace in DataFrame
for state in replace_dest_states:
    filtered_df.DEST_STATE_NM = filtered_df.DEST_STATE_NM.replace(state,"Other")


# Check to make sure binning was successful
filtered_df.DEST_STATE_NM.value_counts()
```

```python
# Create the OneHotEncoder instance
from sklearn.preprocessing import OneHotEncoder
enc2= OneHotEncoder(sparse=False)

#fit the encoder and the produced encoded DataFrame
encoded_plane_dest_df=pd.DataFrame(enc2.fit_transform(filtered_df.DEST_STATE_NM.values.reshape(-1,1)))

#Rename encoded Columns
encoded_plane_dest_df.columns=enc.get_feature_names(['DEST_STATE_NM'])
encoded_plane_dest_df.head(25)
```

```python
encoded_plane_df.iloc[0:50].to_csv('encoded_orgin_state.csv', index=True, header=True)
```

```python
encoded_plane_dest_df.iloc[0:50].to_csv('encoded_dest_state.csv', index=True, header=True)
```
