```python
import pandas as pd
```

```python
flights_df = pd.read_csv("./data/jantojun2020.csv")
flights_df
```

```python
filtered_df= flights_df.drop(['YEAR', 'QUARTER', 'TAIL_NUM','ORIGIN','ORIGIN_CITY_NAME','ORIGIN_STATE_ABR', 'DEST', 'DEST_CITY_NAME','DEST_STATE_ABR', 'CRS_DEP_TIME', 'DEP_TIME','DEP_DELAY_NEW','DEP_DEL15','DEP_DELAY_GROUP','DEP_TIME_BLK','TAXI_OUT','WHEELS_OFF','WHEELS_ON','TAXI_IN','CRS_ARR_TIME','ARR_TIME', 'FL_DATE','CRS_ELAPSED_TIME','ACTUAL_ELAPSED_TIME','AIR_TIME', 'DISTANCE_GROUP', 'ARR_TIME_BLK'],axis=1)
filtered_df
```

```python
carrier_count = filtered_df.MKT_UNIQUE_CARRIER.value_counts()
carrier_count
```

```python
carrier_df = pd.DataFrame(carrier_count)
carrier_df
```

```python
carrier_df['percentage']=(carrier_df['MKT_UNIQUE_CARRIER']/carrier_df['MKT_UNIQUE_CARRIER'].sum())*100
carrier_df
```

```python
delay_df = filtered_df[['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]
delay_df
```

```python
sum = delay_df.sum(skipna = True)
sum
```

```python
sum_df = pd.DataFrame(sum, columns = ["minutes"])
sum_df
```

```python
sum_df['percentage']=(sum_df['minutes']/sum_df['minutes'].sum())*100
sum_df
```
